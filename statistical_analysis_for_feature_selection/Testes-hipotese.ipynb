```python
import numpy as np
from scipy import stats
from numpy.random  import randn, normal
import seed
from scipy.stats import ttest_1samp, ttest_ind, norm
from matplotlib import pyplot as plt
```

```python

seed=(1)
sample =normal(150,10,20)
print('Sample: ', sample)
```

```python
res = ttest_1samp(sample, popmean=155, alternative = 'greater')
print('T-statistic value: ',res.statistic)
print('P-Value: ', res.pvalue)
print('Confidence-interval: ', res.confidence_interval)
```

```python
res.confidence_interval()
```

```python
# seed the random number generator
seed=(1)                # create two independent sample groups
sample1= normal(30, 16, 50)
sample2=normal(33, 18, 50)
print('Sample 1: ',sample1)
print('Sample 2: ',sample2)
```

```python
rng = np.random.default_rng()
```

```python
rvs1 = norm.rvs(loc=5, scale=10, size=500, random_state= rng)
rvs2 = norm.rvs(loc=5, scale=10, size=500, random_state= rng)
stats.ttest_ind(rvs1, rvs2), stats.ttest_ind(rvs1, rvs2, equal_var=False)
```
